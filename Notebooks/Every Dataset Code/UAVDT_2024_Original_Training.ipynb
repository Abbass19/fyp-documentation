```python
import os

def convert_visdrone_to_yolo(visdrone_annotations_dir, yolo_labels_dir, image_width, image_height):
    """
    Convert VisDrone annotations to YOLOv8 format.

    Parameters:
        visdrone_annotations_dir (str): Path to the directory containing VisDrone annotations.
        yolo_labels_dir (str): Path to the directory where YOLO annotations will be saved.
        image_width (int): Width of the images (for normalization).
        image_height (int): Height of the images (for normalization).

    Returns:
        None
    """

    # Ensure the YOLO labels directory exists
    os.makedirs(yolo_labels_dir, exist_ok=True)

    # List all annotation files in the VisDrone annotations directory
    visdrone_files = [f for f in os.listdir(visdrone_annotations_dir) if f.endswith('.txt')]

    # Initialize counters for consistency checks
    total_annotations = 0
    missing_annotations = 0

    for visdrone_file in visdrone_files:
        visdrone_file_path = os.path.join(visdrone_annotations_dir, visdrone_file)

        # Open the VisDrone annotation file
        with open(visdrone_file_path, 'r') as f:
            lines = f.readlines()

        # Prepare to write YOLO annotations
        yolo_annotations = []

        for line in lines:
            parts = line.strip().split(',')

            if len(parts) < 6:
                continue  # Skip invalid lines

            # Parse VisDrone annotation fields
            bbox_left, bbox_top, bbox_width, bbox_height, score, object_category = map(float, parts[:6])

            # Convert to YOLO format
            center_x = (bbox_left + bbox_width / 2) / image_width
            center_y = (bbox_top + bbox_height / 2) / image_height
            width = bbox_width / image_width
            height = bbox_height / image_height

            # The object_category is assumed to be the class ID
            class_id = int(object_category)  # Class ID is an integer in YOLO

            # Append the YOLO formatted annotation
            yolo_annotations.append(f"{class_id} {center_x} {center_y} {width} {height}")

        # Check if there are annotations to save
        if yolo_annotations:
            # Save YOLO annotations to file
            yolo_file_path = os.path.join(yolo_labels_dir, visdrone_file)
            with open(yolo_file_path, 'w') as f:
                f.write("\n".join(yolo_annotations))

            total_annotations += 1
        else:
            missing_annotations += 1

    # Summary of consistency check
    print(f"Total annotations processed: {total_annotations}")
    print(f"Missing or invalid annotations: {missing_annotations}")

    # Check if any annotation files are missing
    if missing_annotations > 0:
        print("Some annotation files were missing or invalid.")
    else:
        print("All annotations were successfully converted.")
```

```python
import os
from PIL import Image

def get_image_dimensions(image_path):
    """
    Get the dimensions of an image using PIL.

    Parameters:
        image_path (str): Path to the image file.

    Returns:
        tuple: (width, height) of the image.
    """
    with Image.open(image_path) as img:
        return img.size  # (width, height)

def convert_visdrone_to_yolo(visdrone_annotations_dir, yolo_labels_dir, image_dir):
    """
    Convert VisDrone annotations to YOLOv8 format based on image dimensions.

    Parameters:
        visdrone_annotations_dir (str): Path to the directory containing VisDrone annotations.
        yolo_labels_dir (str): Path to the directory where YOLO annotations will be saved.
        image_dir (str): Path to the directory containing images (for getting image width/height).

    Returns:
        None
    """
    # Ensure the YOLO labels directory exists
    os.makedirs(yolo_labels_dir, exist_ok=True)

    # List all annotation files in the VisDrone annotations directory
    visdrone_files = [f for f in os.listdir(visdrone_annotations_dir) if f.endswith('.txt')]

    # Initialize counters for consistency checks
    total_annotations = 0
    missing_annotations = 0

    for visdrone_file in visdrone_files:
        visdrone_file_path = os.path.join(visdrone_annotations_dir, visdrone_file)

        # Get corresponding image path
        image_file = visdrone_file.replace('.txt', '.jpg')  # Assuming image is jpg format
        image_file_path = os.path.join(image_dir, image_file)

        # Get image dimensions (width, height)
        image_width, image_height = get_image_dimensions(image_file_path)

        # Open the VisDrone annotation file
        with open(visdrone_file_path, 'r') as f:
            lines = f.readlines()

        # Prepare to write YOLO annotations
        yolo_annotations = []

        for line in lines:
            parts = line.strip().split(',')

            if len(parts) < 6:
                continue  # Skip invalid lines

            # Parse VisDrone annotation fields
            bbox_left, bbox_top, bbox_width, bbox_height, score, object_category = map(float, parts[:6])

            # Convert to YOLO format
            center_x = (bbox_left + bbox_width / 2) / image_width
            center_y = (bbox_top + bbox_height / 2) / image_height
            width = bbox_width / image_width
            height = bbox_height / image_height

            # The object_category is assumed to be the class ID
            class_id = int(object_category)  # Class ID is an integer in YOLO

            # Append the YOLO formatted annotation
            yolo_annotations.append(f"{class_id} {center_x} {center_y} {width} {height}")

        # Check if there are annotations to save
        if yolo_annotations:
            # Save YOLO annotations to file
            yolo_file_path = os.path.join(yolo_labels_dir, visdrone_file)
            with open(yolo_file_path, 'w') as f:
                f.write("\n".join(yolo_annotations))

            total_annotations += 1
        else:
            missing_annotations += 1

    # Summary of consistency check
    print(f"Total annotations processed: {total_annotations}")
    print(f"Missing or invalid annotations: {missing_annotations}")

    # Check if any annotation files are missing
    if missing_annotations > 0:
        print("Some annotation files were missing or invalid.")
    else:
        print("All annotations were successfully converted.")

def convert_all_annotations_to_yolo(dataset_dir, temp_labels_dir):
    """
    Iterate through a list of directories and convert VisDrone annotations to YOLOv8 format for each folder.

    Parameters:
        dataset_dir (str): Path to the main directory containing the folders like M101, M102, etc.
        temp_labels_dir (str): Path to the directory where YOLO annotations will be stored.

    Returns:
        None
    """
    # Ensure the YOLO labels directory exists
    os.makedirs(temp_labels_dir, exist_ok=True)

    # List all directories (e.g., M101, M102, etc.)
    folders = [f for f in os.listdir(dataset_dir) if os.path.isdir(os.path.join(dataset_dir, f))]

    for folder in folders:
        folder_path = os.path.join(dataset_dir, folder)

        # Define the VisDrone annotations directory and images directory for the current folder
        visdrone_annotations_dir = os.path.join(folder_path, 'annotations')
        image_dir = os.path.join(folder_path, 'images')

        # Call the conversion function for each folder
        convert_visdrone_to_yolo(visdrone_annotations_dir, temp_labels_dir, image_dir)

# Example usage
dataset_dir = '/content/UAVDT-2024/UAVDT-2024'  # Path to your UAVDT dataset
temp_labels_dir = '/content/Temp_Dataset/labels'  # Path to save YOLO annotations

# Convert all annotations from the directories
convert_all_annotations_to_yolo(dataset_dir, temp_labels_dir)
```

```python
import os

def unify_annotations(original_annotations_dir, unified_annotations_dir):
    """
    Unify the class labels of the original annotations to match the custom dataset's classes
    based on the transformation table.

    Parameters:
        original_annotations_dir (str): Path to the directory containing original annotations.
        unified_annotations_dir (str): Path to the directory where unified annotations will be saved.

    Returns:
        None
    """
    # Ensure the unified annotations directory exists
    os.makedirs(unified_annotations_dir, exist_ok=True)

    # Define the transformation table (original class -> unified class)
    transformation_table = {
        0: 1,  # small-car -> Car
        1: 2,  # van -> Bus
        2: 2,  # bus -> Bus
        3: 1,  # SUV -> Car
        4: None,  # storage-tank -> delete
        5: None,  # swimming-pool -> delete
        6: None,  # vehicle -> Car
        7: None   # wind-mill -> delete
    }

    # List all annotation files in the original annotations directory
    annotation_files = [f for f in os.listdir(original_annotations_dir) if f.endswith('.txt')]

    # Iterate over each annotation file and process it
    for annotation_file in annotation_files:
        annotation_file_path = os.path.join(original_annotations_dir, annotation_file)

        # Read the original annotation file
        with open(annotation_file_path, 'r') as f:
            lines = f.readlines()

        # Prepare the unified annotations
        unified_annotations = []

        for line in lines:
            parts = line.strip().split()

            if len(parts) != 5:
                continue  # Skip invalid lines

            class_id = int(parts[0])  # Original class id
            # Check if the class_id should be unified or deleted
            if class_id in transformation_table:
                new_class_id = transformation_table[class_id]

                if new_class_id is None:
                    continue  # Skip annotations that need to be deleted

                # Update the class_id in the annotation
                new_annotation = f"{new_class_id} " + " ".join(parts[1:])
                unified_annotations.append(new_annotation)

        # If there are valid annotations, save them
        if unified_annotations:
            unified_annotation_file_path = os.path.join(unified_annotations_dir, annotation_file)
            with open(unified_annotation_file_path, 'w') as f:
                f.write("\n".join(unified_annotations))

    print(f"Unified annotations are saved in {unified_annotations_dir}")

# Example usage
original_annotations_dir = '/content/Temp_Dataset/labels'  # Path to original annotations
unified_annotations_dir = '/content/Temp_Dataset/unified_labels'  # Path to save unified annotations

# Unify annotations
unify_annotations(original_annotations_dir, unified_annotations_dir)
```

```python
import os

def remove_images_without_labels(image_dir, label_dir):
    """
    Remove images that do not have corresponding label files.

    Parameters:
        image_dir (str): Path to the directory containing images.
        label_dir (str): Path to the directory containing label files.

    Returns:
        None
    """
    # List of all image files in the image directory
    image_files = {f.replace('.jpg', ''): f for f in os.listdir(image_dir) if f.endswith('.jpg')}

    # List of all label files in the label directory
    label_files = {f.replace('.txt', ''): f for f in os.listdir(label_dir) if f.endswith('.txt')}

    # Find images without corresponding labels
    missing_labels_images = [image_name for image_name in image_files if image_name not in label_files]

    # Remove images without corresponding labels
    for image_name in missing_labels_images:
        image_file = image_files[image_name]
        image_path = os.path.join(image_dir, image_file)

        if os.path.exists(image_path):
            os.remove(image_path)
            print(f"Deleted image: {image_file}")
        else:
            print(f"Image not found: {image_file}")

# Example usage
image_dir = '/content/Temp_Dataset/images'  # Path to the images directory
label_dir = '/content/Temp_Dataset/labels'  # Path to the labels directory

# Remove images without corresponding labels
remove_images_without_labels(image_dir, label_dir)
```

```python
import os

# Function to calculate number of pictures in each folder and how many pictures to take
def calculate_images_to_select(dataset_dir, target_total_images):
    folder_info = {}
    total_images = 0

    # Loop through all folders like M101, M102, etc.
    for folder in os.listdir(dataset_dir):
        folder_path = os.path.join(dataset_dir, folder)

        if os.path.isdir(folder_path):
            # Define image directory
            images_dir = os.path.join(folder_path, 'images')

            # Get all the image files
            image_files = sorted(os.listdir(images_dir))

            # Count the number of images in this folder
            num_images_in_folder = len(image_files)
            folder_info[folder] = num_images_in_folder

            # Add the number of images to total count
            total_images += num_images_in_folder

    # Now calculate how many images to take from each folder
    images_to_select = {}
    for folder, num_images in folder_info.items():
        # Calculate the proportion of images to select from this folder
        proportion = num_images / total_images
        images_to_take = int(proportion * target_total_images)

        images_to_select[folder] = images_to_take

    # Calculate the total selected images (should match target_total_images)
    total_selected = sum(images_to_select.values())

    return folder_info, images_to_select, total_images, total_selected

# Example usage:
dataset_dir = '/content/UAVDT-2024/UAVDT-2024'  # Path to your dataset
target_total_images = 12548  # Total images you want to select

folder_info, images_to_select, total_images, total_selected = calculate_images_to_select(dataset_dir, target_total_images)

# Display the results
print(f"Total images in the dataset: {total_images}")
print(f"Total images to select: {total_selected}")
print(f"Images to take from each folder:")
for folder, num_images in images_to_select.items():
    # Display the desired output format
    print(f"This directory has {folder_info[folder]} images, we will take {num_images} images.")
```

```python
import os
import random
import shutil

# Function to copy selected images and annotations to the new directory
def select_and_copy_images(dataset_dir, target_total_images, temp_customised_dir, labels_dir):
    # Create required directories if they don't exist
    temp_images_dir = os.path.join(temp_customised_dir, 'images')
    temp_labels_dir = os.path.join(temp_customised_dir, 'labels')

    os.makedirs(temp_images_dir, exist_ok=True)
    os.makedirs(temp_labels_dir, exist_ok=True)

    # Calculate the number of images to take from each folder
    folder_info, images_to_select, total_images, total_selected = calculate_images_to_select(dataset_dir, target_total_images)

    # Copy images and annotations to temp_customised
    total_copied = 0

    for folder, num_images_to_take in images_to_select.items():
        folder_path = os.path.join(dataset_dir, folder)
        images_dir = os.path.join(folder_path, 'images')

        # Get all the image files in the directory
        image_files = sorted(os.listdir(images_dir))

        # Ensure to select non-consecutive images
        selected_images = []
        while len(selected_images) < num_images_to_take:
            # Randomly select an image from the list of available images
            image = random.choice(image_files)
            image_index = image_files.index(image)

            # Avoid consecutive selections
            if image_index not in selected_images and (image_index - 1) not in selected_images and (image_index + 1) not in selected_images:
                selected_images.append(image_index)

        # Copy the selected images and their corresponding annotations
        for index in selected_images:
            image_file = image_files[index]
            image_path = os.path.join(images_dir, image_file)

            # Find the annotation file in the unified_labels directory
            annotation_file = image_file.replace('.jpg', '.txt').replace('.png', '.txt').replace('.jpeg', '.txt')
            annotation_path = os.path.join(labels_dir, annotation_file)

            if os.path.exists(annotation_path):
                # Copy image
                shutil.copy(image_path, temp_images_dir)

                # Copy annotation
                shutil.copy(annotation_path, temp_labels_dir)

                total_copied += 1
            else:
                print(f"Annotation file not found for {image_file}")

    print(f"Total images copied to {temp_images_dir}: {total_copied}")
    return total_copied

# Example usage:
dataset_dir = '/content/UAVDT-2024/UAVDT-2024'  # Path to your dataset
target_total_images = 12548  # Total images you want to select
temp_customised_dir = '/content/temp_customised'  # Path to the custom directory
labels_dir = '/content/Temp_Dataset/unified_labels'  # Path to annotations (flat structure)

# Call the function to select and copy images
total_copied = select_and_copy_images(dataset_dir, target_total_images, temp_customised_dir, labels_dir)

```

```python
import os
import random
import shutil

# Paths
source_images_dir = '/content/temp_customised/images'
source_labels_dir = '/content/temp_customised/labels'
output_dir = '/content/Customised_Dataset_2'

# Subdirectories for train, test, and val
output_images_dir = os.path.join(output_dir, 'images')
output_labels_dir = os.path.join(output_dir, 'labels')

# Create necessary subdirectories
os.makedirs(os.path.join(output_images_dir, 'train'), exist_ok=True)
os.makedirs(os.path.join(output_images_dir, 'test'), exist_ok=True)
os.makedirs(os.path.join(output_images_dir, 'val'), exist_ok=True)

os.makedirs(os.path.join(output_labels_dir, 'train'), exist_ok=True)
os.makedirs(os.path.join(output_labels_dir, 'test'), exist_ok=True)
os.makedirs(os.path.join(output_labels_dir, 'val'), exist_ok=True)

# Get all image files in the source directory
image_files = sorted(os.listdir(source_images_dir))
label_files = sorted(os.listdir(source_labels_dir))

# Ensure that the number of image files matches the number of label files
assert len(image_files) == len(label_files), "Mismatch between image and label files"

# Create a list of (image, label) pairs
data = list(zip(image_files, label_files))

# Shuffle data to ensure randomness
random.shuffle(data)

# Split the data (80% train, 10% test, 10% validation)
train_split = int(0.8 * len(data))
test_split = int(0.1 * len(data))  # 10% for test
val_split = int(0.1 * len(data))  # 10% for validation

# Data splits
train_data = data[:train_split]
test_data = data[train_split:train_split + test_split]
val_data = data[train_split + test_split:]

# Function to copy files
def copy_files(data_split, target_images_dir, target_labels_dir):
    for image_file, label_file in data_split:
        # Define source and target paths
        image_src_path = os.path.join(source_images_dir, image_file)
        label_src_path = os.path.join(source_labels_dir, label_file)

        # Define the target directory (train, test, or val)
        if image_file in [x[0] for x in train_data]:
            image_target_dir = os.path.join(target_images_dir, 'train')
            label_target_dir = os.path.join(target_labels_dir, 'train')
        elif image_file in [x[0] for x in test_data]:
            image_target_dir = os.path.join(target_images_dir, 'test')
            label_target_dir = os.path.join(target_labels_dir, 'test')
        else:
            image_target_dir = os.path.join(target_images_dir, 'val')
            label_target_dir = os.path.join(target_labels_dir, 'val')

        # Copy image and label
        shutil.copy(image_src_path, image_target_dir)
        shutil.copy(label_src_path, label_target_dir)

# Copy the data into the appropriate directories
copy_files(train_data, output_images_dir, output_labels_dir)
copy_files(test_data, output_images_dir, output_labels_dir)
copy_files(val_data, output_images_dir, output_labels_dir)

# Report the number of images in each folder
print(f"Training images: {len(train_data)}")
print(f"Test images: {len(test_data)}")
print(f"Validation images: {len(val_data)}")
```

```python
import os

# Function to count the number of images in each folder
def count_images_in_folders(base_dir):
    # Directories for train, test, and val
    train_dir = os.path.join(base_dir, 'train')
    test_dir = os.path.join(base_dir, 'test')
    val_dir = os.path.join(base_dir, 'val')

    # Count the number of image files in each folder
    train_images = len([f for f in os.listdir(train_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
    test_images = len([f for f in os.listdir(test_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])
    val_images = len([f for f in os.listdir(val_dir) if f.lower().endswith(('.jpg', '.jpeg', '.png'))])

    return train_images, test_images, val_images

# Specify the base directory
base_directory = '/content/Customised_Dataset_2/images'

# Get the count of images in each folder
train_count, test_count, val_count = count_images_in_folders(base_directory)

# Print the results
print(f"Number of images in train: {train_count}")
print(f"Number of images in test: {test_count}")
print(f"Number of images in val: {val_count}")
```

```python
import os
import shutil

# Directories
source_images_dir = '/content/Customised_Dataset_2/images'
target_images_dir = '/content/Non_Unified_Labels/images'

# Ensure that the target directory exists
os.makedirs(target_images_dir, exist_ok=True)

# Loop through each subfolder (train, test, val) inside the source images directory
for subfolder in ['train', 'test', 'val']:  # Loop through specific subfolders
    source_subfolder_path = os.path.join(source_images_dir, subfolder)
    target_subfolder_path = os.path.join(target_images_dir, subfolder)

    # Ensure that the target subfolder exists
    os.makedirs(target_subfolder_path, exist_ok=True)

    # Loop through all images in the subfolder
    for image_file in os.listdir(source_subfolder_path):
        # Check if it's a valid image file (jpg, jpeg, png)
        if image_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            source_image_path = os.path.join(source_subfolder_path, image_file)
            target_image_path = os.path.join(target_subfolder_path, image_file)

            # Copy the image to the target subfolder
            shutil.copy(source_image_path, target_image_path)
            print(f"Copied {image_file} to {target_subfolder_path}")

print("Image copy process completed.")
```

```python
import os
import shutil

# Directories
source_labels_dir = '/content/Temp_Dataset/labels'
target_labels_dir = '/content/Non_Unified_Labels/labels'

# Ensure that the target label directories exist
os.makedirs(os.path.join(target_labels_dir, 'train'), exist_ok=True)
os.makedirs(os.path.join(target_labels_dir, 'test'), exist_ok=True)
os.makedirs(os.path.join(target_labels_dir, 'val'), exist_ok=True)

# Function to copy the annotation file to the corresponding directory
def copy_annotation_files(image_dir, label_dir):
    # Loop through each image in the directory
    for image_file in os.listdir(image_dir):
        # Ensure the file is an image
        if image_file.lower().endswith(('.jpg', '.jpeg', '.png')):
            # Generate the corresponding annotation file name
            label_file = image_file.replace('.jpg', '.txt').replace('.jpeg', '.txt').replace('.png', '.txt')

            # Source annotation path
            source_label_path = os.path.join(source_labels_dir, label_file)

            # Check if the corresponding annotation file exists
            if os.path.exists(source_label_path):
                # Determine the target path for the annotation file
                if "train" in image_dir:
                    target_label_path = os.path.join(target_labels_dir, 'train', label_file)
                elif "test" in image_dir:
                    target_label_path = os.path.join(target_labels_dir, 'test', label_file)
                elif "val" in image_dir:
                    target_label_path = os.path.join(target_labels_dir, 'val', label_file)

                # Copy the annotation file to the corresponding directory
                shutil.copy(source_label_path, target_label_path)
                print(f"Copied annotation: {label_file} to {target_label_path}")
            else:
                print(f"Annotation file for {image_file} does not exist, skipping.")

# Loop through each subfolder in the images directory (train, test, val)
for subfolder in ['train', 'test', 'val']:
    # Define the image subfolder path
    image_subfolder_path = os.path.join('/content/Non_Unified_Labels/images', subfolder)

    # Copy the corresponding annotation files
    copy_annotation_files(image_subfolder_path, target_labels_dir)

print("Annotation copy process completed.")
```
