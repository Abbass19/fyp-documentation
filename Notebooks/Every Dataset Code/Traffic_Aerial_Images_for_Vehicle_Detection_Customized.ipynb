```python
#Here we need to make a count on directories to delete images accordingly.
# All we want from this dataset 15070 is 6119 image. Should be Distributed Equally
import os
import random

# Define the dataset path
dataset_path = "/content/dataset/dataset"

# Store image counts per folder
folder_image_counts = {}
total_images = 0

# Count images in each subfolder
for subfolder in os.listdir(dataset_path):
    subfolder_path = os.path.join(dataset_path, subfolder)

    if os.path.isdir(subfolder_path):
        images = [f for f in os.listdir(subfolder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        folder_image_counts[subfolder] = len(images)
        total_images += len(images)

# Ensure there are enough images
if total_images < 6120:
    raise ValueError("Not enough images in dataset to select 6120.")

# Determine how many images to take from each folder (initial selection)
selected_images = {}
selection_summary = []
total_selected = 0

for subfolder, count in folder_image_counts.items():
    num_to_select = round((count / total_images) * 6120)  # Proportional selection
    selected_images[subfolder] = num_to_select
    total_selected += num_to_select

# Adjust rounding errors to ensure the total is exactly 6120
difference = 6120 - total_selected  # Difference due to rounding
sorted_folders = sorted(folder_image_counts, key=lambda x: -folder_image_counts[x])  # Sort by image count

# Distribute the difference
for i in range(abs(difference)):
    folder_to_adjust = sorted_folders[i % len(sorted_folders)]
    selected_images[folder_to_adjust] += 1 if difference > 0 else -1

# Verify final count
assert sum(selected_images.values()) == 6120, "Total selected images do not sum to 6120!"

# Store final selection summary
selection_summary = [(folder, selected_images[folder]) for folder in selected_images]

# Print summary
print(f"Total images selected: {sum(selected_images.values())}\n")
print("Selection Summary (Folder -> Images Selected):")
for folder, count in selection_summary:
    print(f"{folder} -> {count}")
```

```python
# A code to know how many pictures are there from each folder. Also how much to take from each one so that the total is 6120. Also how much to delete
# from each folder, with deleting the corresponding annotation. (This code makes a report only)
import os

# Define the dataset path
dataset_path = "/content/dataset/dataset"

# Store image counts per folder
folder_image_counts = {}
total_images = 0

# Count images in each subfolder
for subfolder in os.listdir(dataset_path):
    subfolder_path = os.path.join(dataset_path, subfolder)

    if os.path.isdir(subfolder_path):
        images = [f for f in os.listdir(subfolder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        folder_image_counts[subfolder] = len(images)
        total_images += len(images)

# Calculate excess images
excess_images = total_images - 6120

if excess_images <= 0:
    print("No need to delete images. Storage is sufficient.")
else:
    print(f"Total excess images to delete: {excess_images}")

    # Determine initial deletion count per folder
    deletion_counts = {}
    total_deleted = 0

    for subfolder, count in folder_image_counts.items():
        num_to_delete = round((count / total_images) * excess_images)  # Proportional deletion
        deletion_counts[subfolder] = num_to_delete
        total_deleted += num_to_delete

    # Adjust rounding errors to ensure exact deletion
    difference = excess_images - total_deleted
    sorted_folders = sorted(folder_image_counts, key=lambda x: -folder_image_counts[x])  # Sort by image count

    for i in range(abs(difference)):
        folder_to_adjust = sorted_folders[i % len(sorted_folders)]
        deletion_counts[folder_to_adjust] += 1 if difference > 0 else -1

    # Verify deletion count
    assert sum(deletion_counts.values()) == excess_images, "Total deletions do not sum correctly!"

    # Store final deletion summary
    deletion_summary = [(folder, deletion_counts[folder]) for folder in deletion_counts]

    # Print deletion summary
    print("\nDeletion Summary (Folder -> Images to Delete):")
    for folder, count in deletion_summary:
        print(f"{folder} -> {count}")
```

```python
# This code acyually deletes the images and the remaining are 6120 as variaint as possible. Take into consideration when deleting to delete in a linspace way.
# Which means don't delete consecutive images
import os
import numpy as np

# Define the dataset path
dataset_path = "/content/dataset/dataset"

# Function to delete images and corresponding annotations
def delete_images_linspace(folder, num_to_delete):
    folder_path = os.path.join(dataset_path, folder)
    images = sorted([f for f in os.listdir(folder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])

    if num_to_delete > len(images):
        print(f"Warning: Trying to delete {num_to_delete} images, but only {len(images)} exist in {folder}.")
        num_to_delete = len(images)

    if num_to_delete > 0:
        indices = np.linspace(0, len(images) - 1, num_to_delete, dtype=int)  # Select spread-out indices
        images_to_delete = [images[i] for i in indices]

        for img in images_to_delete:
            img_path = os.path.join(folder_path, img)
            annotation_path = os.path.join(folder_path, img.rsplit(".", 1)[0] + ".txt")  # Check for annotation

            # Delete the image
            os.remove(img_path)

            # Delete the annotation if it exists
            if os.path.exists(annotation_path):
                os.remove(annotation_path)

# Execute deletion based on the calculated `deletion_summary`
for folder, num_to_delete in deletion_summary:
    delete_images_linspace(folder, num_to_delete)

# Re-scan remaining images after deletion
remaining_images = {}
total_remaining = 0

for subfolder in os.listdir(dataset_path):
    subfolder_path = os.path.join(dataset_path, subfolder)

    if os.path.isdir(subfolder_path):
        images = [f for f in os.listdir(subfolder_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        remaining_images[subfolder] = len(images)
        total_remaining += len(images)

# Print remaining images summary
print("\nRemaining Images Summary (Folder -> Images Left):")
for folder, count in remaining_images.items():
    print(f"{folder} -> {count}")

print(f"\nTotal images remaining after deletion: {total_remaining}")
```

```python
import os

def get_folder_size(folder_path):
    total_size = 0
    for dirpath, dirnames, filenames in os.walk(folder_path):
        for file in filenames:
            file_path = os.path.join(dirpath, file)
            total_size += os.path.getsize(file_path)  # Add file size
    return total_size

# Define dataset path
dataset_path = "/content/dataset/dataset"

# Get folder size in bytes
folder_size_bytes = get_folder_size(dataset_path)

# Convert size to MB and GB
folder_size_MB = folder_size_bytes / (1024 * 1024)
folder_size_GB = folder_size_MB / 1024

# Print results
print(f"Folder size of '{dataset_path}':")
print(f"🔹 {folder_size_bytes:,} bytes")
print(f"🔹 {folder_size_MB:.2f} MB")
print(f"🔹 {folder_size_GB:.2f} GB")
```

```python
import os
import shutil

# Define source dataset path
dataset_path = "/content/dataset/dataset"

# Define target paths
target_images_path = "/content/Whole_Data/images"
target_labels_path = "/content/Whole_Data/labels"

# Ensure target directories exist
os.makedirs(target_images_path, exist_ok=True)
os.makedirs(target_labels_path, exist_ok=True)

# Move files from each subfolder
for subfolder in os.listdir(dataset_path):
    subfolder_path = os.path.join(dataset_path, subfolder)

    if os.path.isdir(subfolder_path):
        files = os.listdir(subfolder_path)

        for file in files:
            file_path = os.path.join(subfolder_path, file)

            # Check if it's an image
            if file.lower().endswith(('.png', '.jpg', '.jpeg')):
                shutil.move(file_path, os.path.join(target_images_path, file))

            # Check if it's a label (assuming .txt format)
            elif file.lower().endswith('.txt'):
                shutil.move(file_path, os.path.join(target_labels_path, file))

print("\n✅ All images and labels have been moved successfully!")
shutil.rmtree("/content/dataset")
```

```python
import os

# Define paths
images_path = "/content/Whole_Data/images"
labels_path = "/content/Whole_Data/labels"

# Get sorted list of images (to maintain order)
image_files = sorted([f for f in os.listdir(images_path) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])

# Starting index
start_index = 10000

# Rename images and labels
for idx, image_file in enumerate(image_files):
    # Get old image and label paths
    old_image_path = os.path.join(images_path, image_file)
    old_label_path = os.path.join(labels_path, os.path.splitext(image_file)[0] + ".txt")  # Corresponding label

    # Define new names
    new_filename = str(start_index + idx)  # New base name
    new_image_path = os.path.join(images_path, new_filename + os.path.splitext(image_file)[1])  # Keep original extension
    new_label_path = os.path.join(labels_path, new_filename + ".txt")  # Label in .txt format

    # Rename image
    os.rename(old_image_path, new_image_path)

    # Rename label if it exists
    if os.path.exists(old_label_path):
        os.rename(old_label_path, new_label_path)

print("\n✅ All images and labels have been renamed successfully!")
```

```python
import os

# Define paths for labels
source_labels_path = "/content/Whole_Data/labels"
destination_labels_path = "/content/Whole_Data/unified_labels"

# Define the label transformation mapping
label_mapping = {
    0: 0,
    1: None  # This means labels with class_id 1 will be removed
}

# Function to process and transform the labels
def unify_labels():
    # Ensure the destination folder exists
    os.makedirs(destination_labels_path, exist_ok=True)

    # Process each label file in the source folder
    for label_file in os.listdir(source_labels_path):
        if label_file.endswith(".txt"):
            label_file_path = os.path.join(source_labels_path, label_file)

            # Read the label file
            with open(label_file_path, 'r') as file:
                lines = file.readlines()

            # Create a list to store the transformed lines
            new_lines = []

            # Process each line in the label file
            for line in lines:
                parts = line.strip().split()
                class_id = int(parts[0])  # Class ID (0, 1, etc.)

                # Transform based on label_mapping
                if class_id in label_mapping:
                    new_class_id = label_mapping[class_id]

                    # If the new class ID is not None, keep the entry
                    if new_class_id is not None:
                        new_line = f"{new_class_id} {' '.join(parts[1:])}\n"
                        new_lines.append(new_line)

            # Write the transformed labels to the new destination
            dest_label_file_path = os.path.join(destination_labels_path, label_file)
            with open(dest_label_file_path, 'w') as dest_file:
                dest_file.writelines(new_lines)

    print("\n✅ Label unification completed successfully!")

# Call the function to unify the labels
unify_labels()
```

```python
#Spliting Data
import os
import shutil
import random

# Set seed for reproducibility
random.seed(42)

# Define source paths
source_images = "/content/Whole_Data/images"
source_labels = "/content/Whole_Data/labels"
source_unified_labels = "/content/Whole_Data/unified_labels"

# Define destination paths
dest_images = "/content/Unified_Dataset/images"
dest_labels = "/content/Original_Dataset/labels"
dest_unified_labels = "/content/Unified_Dataset/labels"

# Define train-test-val split ratios
train_ratio = 0.7
test_ratio = 0.2
val_ratio = 0.1

# Get all images (sorted for consistency)
image_files = sorted([f for f in os.listdir(source_images) if f.lower().endswith(('.png', '.jpg', '.jpeg'))])

# Shuffle for randomness
random.shuffle(image_files)

# Compute split indices
total_images = len(image_files)
train_end = int(train_ratio * total_images)
test_end = train_end + int(test_ratio * total_images)

# Split data
train_files = image_files[:train_end]
test_files = image_files[train_end:test_end]
val_files = image_files[test_end:]

# Function to move files
def move_files(file_list, source_folder, destination_folder):
    os.makedirs(destination_folder, exist_ok=True)
    for file in file_list:
        src = os.path.join(source_folder, file)
        dst = os.path.join(destination_folder, file)
        if os.path.exists(src):
            shutil.move(src, dst)

# Move images
move_files(train_files, source_images, os.path.join(dest_images, "train"))
move_files(test_files, source_images, os.path.join(dest_images, "test"))
move_files(val_files, source_images, os.path.join(dest_images, "val"))

# Move corresponding labels
move_files([f.replace(os.path.splitext(f)[1], ".txt") for f in train_files], source_labels, os.path.join(dest_labels, "train"))
move_files([f.replace(os.path.splitext(f)[1], ".txt") for f in test_files], source_labels, os.path.join(dest_labels, "test"))
move_files([f.replace(os.path.splitext(f)[1], ".txt") for f in val_files], source_labels, os.path.join(dest_labels, "val"))

# Move corresponding unified labels
move_files([f.replace(os.path.splitext(f)[1], ".txt") for f in train_files], source_unified_labels, os.path.join(dest_unified_labels, "train"))
move_files([f.replace(os.path.splitext(f)[1], ".txt") for f in test_files], source_unified_labels, os.path.join(dest_unified_labels, "test"))
move_files([f.replace(os.path.splitext(f)[1], ".txt") for f in val_files], source_unified_labels, os.path.join(dest_unified_labels, "val"))

print("\n✅ Splitting and moving completed successfully!")
shutil.rmtree("/content/Whole_Data")

```
