```python
import os
import shutil

# Define the dataset path
dataset_path = "/content/aerial_traffic_dataset"
aerial_images_path = f"{dataset_path}/aerial_images"

# Step 1: Create the correct folder structure
folders = [
    f"{aerial_images_path}/images/train",
    f"{aerial_images_path}/images/valid",
    f"{aerial_images_path}/images/test",
    f"{aerial_images_path}/label/train",
    f"{aerial_images_path}/label/valid",
    f"{aerial_images_path}/label/test"
]

for folder in folders:
    os.makedirs(folder, exist_ok=True)

# Step 2: Function to move images and labels safely
def move_files(source_images, source_labels, target_images, target_labels):
    if os.path.exists(source_images):
        for file in os.listdir(source_images):
            if file.endswith((".jpg", ".png", ".jpeg")):  # Move only image files
                shutil.move(os.path.join(source_images, file), os.path.join(target_images, file))

    if os.path.exists(source_labels):
        for file in os.listdir(source_labels):
            if file.endswith(".txt"):  # Move only annotation files
                shutil.move(os.path.join(source_labels, file), os.path.join(target_labels, file))

# Step 3: Move train, valid, and test data to the correct locations
move_files(f"{dataset_path}/train/images", f"{dataset_path}/train/labels", f"{aerial_images_path}/images/train", f"{aerial_images_path}/label/train")
move_files(f"{dataset_path}/valid/images", f"{dataset_path}/valid/labels", f"{aerial_images_path}/images/valid", f"{aerial_images_path}/label/valid")
move_files(f"{dataset_path}/test/images", f"{dataset_path}/test/labels", f"{aerial_images_path}/images/test", f"{aerial_images_path}/label/test")

print(" Dataset structure is now correct!")
```

```python
import os

def unify_aerial_labels(input_dir, output_dir):
    """
    Reads YOLO format annotation files from the Aerial Traffic dataset, transforms class labels
    to match the custom dataset, and saves the updated labels to the output directory.

    Args:
        input_dir (str): Path to the directory containing original Aerial Traffic annotations.
        output_dir (str): Path to save the transformed annotations.
    """

    # Transformation mapping based on the provided table
    label_mapping = {
        0: 2,    # PMT -> Truck
        1: 1,    # articulated-bus -> Bus
        2: 1,    # bus -> Bus
        3: 0,    # car -> Car
        4: None, # freight -> DELETE
        5: None, # motorbike -> DELETE
        6: 1,    # small bus -> Bus
        7: 2     # truck -> Truck
    }

    # Ensure output directory exists
    os.makedirs(output_dir, exist_ok=True)

    file_count = 0  # Track the number of processed files

    # Process each annotation file in the input directory
    for filename in os.listdir(input_dir):
        if filename.endswith(".txt"):  # Process only annotation files
            input_file_path = os.path.join(input_dir, filename)
            output_file_path = os.path.join(output_dir, filename)

            with open(input_file_path, "r") as infile, open(output_file_path, "w") as outfile:
                for line in infile:
                    parts = line.strip().split()
                    if len(parts) < 5:
                        continue  # Skip invalid lines

                    class_id = int(parts[0])  # Extract class label
                    if class_id in label_mapping:
                        new_class = label_mapping[class_id]
                        if new_class is not None:
                            # Write the transformed annotation
                            outfile.write(f"{new_class} {' '.join(parts[1:])}\n")

            file_count += 1  # Increment renamed file count

    print(f" Label unification complete! {file_count} annotation files transformed and saved in: {output_dir}")
```

```python
import os

def rename_images_and_annotations(original_images_dir, original_labels_dir, unified_images_dir, unified_labels_dir, start_index=10000):
    """
    Renames image and annotation files in 'train', 'valid', and 'test' folders for both
    the original and unified datasets, ensuring corresponding pairs are renamed consistently.

    Args:
        original_images_dir (str): Path to the original images directory (root containing train/valid/test).
        original_labels_dir (str): Path to the original labels directory (root containing train/valid/test).
        unified_images_dir (str): Path to the unified images directory (root containing train/valid/test).
        unified_labels_dir (str): Path to the unified labels directory (root containing train/valid/test).
        start_index (int): Starting number for renaming (default: 10000).
    """

    # Define subdirectories to process
    subsets = ["train", "valid", "test"]

    file_count = 0  # Track the total number of renamed files

    for subset in subsets:
        orig_img_path = os.path.join(original_images_dir, subset)
        orig_lbl_path = os.path.join(original_labels_dir, subset)
        unif_img_path = os.path.join(unified_images_dir, subset)
        unif_lbl_path = os.path.join(unified_labels_dir, subset)

        # Get a sorted list of image files in the subset (assumes all have annotations)
        image_files = sorted([f for f in os.listdir(orig_img_path) if f.endswith(('.jpg', '.png', '.jpeg'))])

        for i, image_file in enumerate(image_files):
            base_name, ext = os.path.splitext(image_file)  # Extract name and extension
            annotation_file = f"{base_name}.txt"  # Corresponding annotation filename

            # Generate new filename
            new_name = f"{start_index + file_count}{ext}"  # Image name
            new_annotation_name = f"{start_index + file_count}.txt"  # Annotation name

            # Define full file paths
            orig_img_file = os.path.join(orig_img_path, image_file)
            orig_lbl_file = os.path.join(orig_lbl_path, annotation_file)
            unif_img_file = os.path.join(unif_img_path, image_file)
            unif_lbl_file = os.path.join(unif_lbl_path, annotation_file)

            # Define new file paths
            new_orig_img_file = os.path.join(orig_img_path, new_name)
            new_orig_lbl_file = os.path.join(orig_lbl_path, new_annotation_name)
            new_unif_img_file = os.path.join(unif_img_path, new_name)
            new_unif_lbl_file = os.path.join(unif_lbl_path, new_annotation_name)

            # Rename files in the original dataset
            if os.path.exists(orig_img_file):
                os.rename(orig_img_file, new_orig_img_file)
            if os.path.exists(orig_lbl_file):
                os.rename(orig_lbl_file, new_orig_lbl_file)

            # Rename files in the unified dataset
            if os.path.exists(unif_img_file):
                os.rename(unif_img_file, new_unif_img_file)
            if os.path.exists(unif_lbl_file):
                os.rename(unif_lbl_file, new_unif_lbl_file)

            file_count += 1  # Increment total renamed file count

    print(f" Renamed {file_count} images and annotations across train, valid, and test, starting from {start_index}.")
```

```python
import yaml
import os

def create_yaml_files():
    # Define file paths
    original_yaml_path = "/content/aerial_images_original/Aerial.yaml"
    unified_yaml_path = "/content/aerial_images_unified/Aerial.yaml"

    # Data for aerial_images_original
    original_data = {
        "train": "/content/aerial_images_original/images/train",
        "val": "/content/aerial_images_original/images/valid",
        "test": "/content/aerial_images_original/images/test",
        "nc": 8,
        "names": {
            0: "PMT",
            1: "articulated-bus",
            2: "bus",
            3: "car",
            4: "freight",
            5: "motorbike",
            6: "small bus",
            7: "truck"
        }
    }

    # Data for aerial_images_unified
    unified_data = {
        "train": "/content/aerial_images_unified/images/train",
        "val": "/content/aerial_images_unified/images/valid",
        "test": "/content/aerial_images_unified/images/test",
        "nc": 11,
        "names": {
            0: "Car",
            1: "Bus",
            2: "Truck"
        }
    }

    # Ensure directories exist
    os.makedirs(os.path.dirname(original_yaml_path), exist_ok=True)
    os.makedirs(os.path.dirname(unified_yaml_path), exist_ok=True)

    # Save aerial.yaml for original dataset
    with open(original_yaml_path, "w") as file:
        yaml.dump(original_data, file, default_flow_style=False)

    # Save aerial.yaml for unified dataset
    with open(unified_yaml_path, "w") as file:
        yaml.dump(unified_data, file, default_flow_style=False)

    print(f" YAML files created successfully!\n - {original_yaml_path}\n - {unified_yaml_path}")

# Run the function
create_yaml_files()
```

```python
import shutil
import os

def zip_and_save_to_drive(directory_path, drive_folder="/content/drive/MyDrive/"):
    """
    Compresses the given directory into a zip file and saves it to Google Drive.

    Args:
        directory_path (str): The full path of the directory to be zipped.
        drive_folder (str): Path to the Google Drive folder to save the zip file.

    Returns:
        str: The path of the created zip file in Google Drive.
    """
    if not os.path.exists(directory_path):
        print(f" Warning: Directory not found: {directory_path}")
        return None

    # Extract directory name and define zip file path in Google Drive
    dir_name = os.path.basename(directory_path)
    zip_file_path = os.path.join(drive_folder, f"{dir_name}.zip")

    # Create a zip file
    shutil.make_archive(zip_file_path.replace(".zip", ""), 'zip', directory_path)
    print(f" Successfully saved zip file to Google Drive: {zip_file_path}")

    return zip_file_path  # Return the zip file path in Google Drive

# Example usage:
zip_and_save_to_drive("/content/aerial_images_original")
zip_and_save_to_drive("/content/aerial_images_unified")
```
